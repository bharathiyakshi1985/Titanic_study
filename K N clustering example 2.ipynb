```python
#Import required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
dataset=pd.read_csv('C:/Users/Bharathi/Downloads/xclara.csv')
dataset.shape
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
f1=dataset['V1'].values
f2=dataset['V2'].values
X=np.array(list(zip(f1,f2)))
plt.scatter(f1,f2,c='blue',s=7)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans=kmeans.fit(X)
labels=kmeans.predict(X)
centroids=kmeans.cluster_centers_
```

```python
pd.DataFrame({"V1":dataset['V1'],"V2":dataset['V2'],"Labels":labels})
```
